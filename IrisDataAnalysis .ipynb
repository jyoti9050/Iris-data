# statstical function 


```python
#complete statistics  
iris.describe()    

print(iris.columns)

#find mean of particular column
iris['SepalLengthCm'].mean()



#find mode of SepalWidthCm column mode 
iris['SepalWidthCm'].mode()
  
#median of complete data set 
iris.median()
 
    
print(iris.median(axis=1))  #find median acording to column means axis=1

```

# group by :used to split the data into groups based on some criteria.

```python
iris.Species.unique()   #unique value of each species 
```

```python
#apply groupby() function to group the data on Species value 
gk=iris.groupby('Species')
#Let's print the first enteries in all the groups formed.
gk.groups    #to see all the groups 

```

```python
#groupby with multiple columns 
print(iris.groupby(['Species','Id']).groups)
```

```python
# select a single group 
gk1=iris.groupby('Species')
print(gk1.get_group('Iris-virginica'))
```

# Aggregations


```python
gk2=iris.groupby('Species')
print(gk2['SepalLengthCm'].agg(np.mean))

#size of each group is applying by size() funtion 

print(gk2.agg(np.size))
```
